```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def conv(s):
    for _ in range(3):
        return int(s, 16)
```

```python
non_struct_data = np.loadtxt('itcm_data_if_curr_exp_stdp.txt', dtype="int32, S20, int32, S20", 
                             converters={0: conv, 2:conv})
```

```python
struct_data = np.loadtxt('itcm_data_if_curr_exp_stdp_struct.txt', dtype="int32, S20, int32, S20", 
                             converters={0: conv, 2:conv})
```

```python
# Two subplots, unpack the axes array immediately
f, ax1 = plt.subplots(1, 1 ,figsize=(16,8))
plt.step(xrange(non_struct_data.size), non_struct_data['f2'])
plt.xticks(xrange(non_struct_data.size), non_struct_data['f3'], rotation=45)
```
