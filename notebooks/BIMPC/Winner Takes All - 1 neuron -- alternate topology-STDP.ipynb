```python
import spynnaker.pyNN as p
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from brian2.units import *
```

```python
p.setup(timestep=1.0, min_delay=1.0, max_delay=144.0)
n_neurons = 1 
```

```python
run_time = 1000 * ms
f_p1 = 50 * Hz
f_p2 = 10 * Hz
```

```python
cell_params_lif = {'cm': 0.25,
                   'i_offset': 0.0,
                   'tau_m': 20.0,
                   'tau_refrac': 2.0,
                   'tau_syn_E': 5.0,
                   'tau_syn_I': 5.0,
                   'v_reset': -70.0,
                   'v_rest': -65.0,
                   'v_thresh': -50.0
                   }
```

```python
weight_to_spike = 2.
delay = 1
```

```python
p1 = p.Population(n_neurons, p.IF_curr_exp, cell_params_lif,
                   label='pop_1')
p2 = p.Population(n_neurons, p.IF_curr_exp, cell_params_lif,
                   label='pop_2')
inhibitor_population = p.Population(n_neurons, p.IF_curr_exp, cell_params_lif,
                   label='inhibitor_pop')
```

```python
stdp_model = p.STDPMechanism(
    timing_dependence=p.SpikePairRule(tau_plus=20., tau_minus=20.0,nearest=True),
    weight_dependence=p.AdditiveWeightDependence(w_min=0, w_max=2.,A_plus=0.02, A_minus=0.02)
)
```

```python
p.Projection(inhibitor_population, p2, p.OneToOneConnector(weight_to_spike/2.), target='inhibitory', 
             label='inhib ->2')
p.Projection(inhibitor_population, p1, p.OneToOneConnector(weight_to_spike/2.), target='inhibitory', 
             label='inhib ->1')
p.Projection(p1, inhibitor_population, p.OneToOneConnector(weight_to_spike*2.), target='excitatory', 
             label='inhib ->1', synapse_dynamics=p.SynapseDynamics(slow=stdp_model))
p.Projection(p2, inhibitor_population, p.OneToOneConnector(weight_to_spike*2.), target='excitatory', 
             label='inhib ->1', synapse_dynamics=p.SynapseDynamics(slow=stdp_model))
```

```python
p.Projection(p1, p1, p.OneToOneConnector(weight_to_spike), label='excite 1->1')
p.Projection(p2, p2, p.OneToOneConnector(weight_to_spike), label='excite 2->2')
p.Projection(inhibitor_population, inhibitor_population, p.OneToOneConnector(weight_to_spike/2.), 
             label='excite inhib->inhib')
```

```python
p1.record_v()
p1.record_gsyn()
p1.record()
p2.record_v()
p2.record_gsyn()
p2.record()
```

```python
# Input rates
p1_spike_times= np.linspace(0, run_time/ms, run_time * f_p1)
p2_spike_times= np.linspace(0, run_time/ms, run_time * f_p2)

t1 = p1_spike_times + run_time/ms
t2 = p2_spike_times + run_time/ms

spikeArray_p1 = {'spike_times': np.concatenate((p1_spike_times,t2))}
spikeArray_p2 = {'spike_times': np.concatenate((p2_spike_times,t1))}

# Input populations
input_p1 = p.Population(1, p.SpikeSourceArray, spikeArray_p1,label='inputSpikes_p1')
input_p2 = p.Population(1, p.SpikeSourceArray, spikeArray_p2,label='inputSpikes_p2')

# Input connections
p.Projection(input_p1, p1, p.OneToOneConnector(weight_to_spike), label='excite input_p1->1')
p.Projection(input_p2, p2, p.OneToOneConnector(weight_to_spike), label='excite input_p2->2')
```

```python
p.run(2 * run_time/ms)
```

```python
v_p1 = p1.get_v(compatible_output=True)
gsyn_p1 = p1.get_gsyn(compatible_output=True)
spikes_p1 = p1.getSpikes(compatible_output=True)
v_p2 = p2.get_v(compatible_output=True)
gsyn_p2 = p2.get_gsyn(compatible_output=True)
spikes_p2 = p2.getSpikes(compatible_output=True)
p.end()
```

```python
if spikes_p1 is not None:
#     print spikes_p1p2
    plt.figure()
    plt.plot([i[1] for i in spikes_p1], [i[0] for i in spikes_p1], ".")
    plt.xlabel('Time/ms')
    plt.ylabel('spikes')
    plt.title('spikes')
```

```python
if spikes_p2 is not None:
#     print spikes_p1
    plt.figure()
    plt.plot([i[1] for i in spikes_p2], [i[0] for i in spikes_p2], ".")
    plt.xlabel('Time/ms')
    plt.ylabel('spikes')
    plt.title('spikes')
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(v_p1[:, 2], c='b')
plt.plot(v_p2[:, 2], c='r')
for t in spikes_p1[:, 1]:
    plt.axvline(t, ls='-', c='b', lw=1, alpha=.5)
for t in spikes_p2[:, 1]:
    plt.axvline(t, ls=':', c='r', lw=1, alpha=.7)
    
plt.axvline(run_time/ms, ls='-', c='g', lw=1, alpha=1)
plt.axvline(2*run_time/ms, ls='-', c='g', lw=1, alpha=1)
plt.xlabel('Time/ms')
plt.ylabel('mV')
plt.title('Voltage')
```
